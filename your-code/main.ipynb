## Step 3: Use list comprehensions with conditions to clean the link list.

There are two types of links, absolute and relative. Absolute links have the full URL and begin with http while relative links begin with a forward slash (/) and point to an internal page within the wikipedia.org domain. Clean the respective types of URLs as follows.

- Absolute Links: Create a list of these and remove any that contain a percentage sign (%).
- Relativel Links: Create a list of these, add the domain to the link so that you have the full URL, and remove any that contain a percentage sign (%).
- Combine the list of absolute and relative links and ensure there are no duplicates.

```python
domain = 'http://wikipedia.org'
```

```python
# URL de la página web
domain = 'http://wikipedia.org'

# Realizar la solicitud GET a la URL
response = requests.get(url)
html_content = response.content

# Crear un objeto BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Encontrar todos los enlaces en la página
links = soup.find_all('a')

# Extraer los enlaces únicos de los atributos 'href'
unique_links = set(link.get('href') for link in links if link.get('href'))

# Filtrar los enlaces que comienzan con 'http' o '/'
filtered_links = [link for link in unique_links if link.startswith('http') or link.startswith('/')]

# Filtrar y limpiar enlaces absolutos
absolute_links = [link for link in filtered_links if link.startswith('http') and '%' not in link]

# Filtrar, limpiar y agregar dominio a los enlaces relativos
relative_links = [domain + link if link.startswith('/') and '%' not in link else link for link in filtered_links]

# Combinar y eliminar duplicados
clean_links = list(set(absolute_links + relative_links))

# Imprimir los enlaces limpios
for link in clean_links:
    print(link)
```

## Step 5: Write a function called index_page that accepts a link and does the following.

- Tries to request the content of the page referenced by that link.
- Slugifies the filename using the `slugify` function from the [python-slugify](https://pypi.org/project/python-slugify/) library and adds a .html file extension.
    - If you don't already have the python-slugify library installed, you can pip install it as follows: `$ pip install python-slugify`.
    - To import the slugify function, you would do the following: `from slugify import slugify`.
    - You can then slugify a link as follows `slugify(link)`.
- Creates a file in the wikipedia folder using the slugified filename and writes the contents of the page to the file.
- If an exception occurs during the process above, just `pass`.

```python
!pip install python-slugify
```

```python
from slugify import slugify

unicode=str()
```

```python
import requests
import slugify

def index_page(link):
    try:
        response = requests.get(link)
        if response.status_code == 200:
            page_content = response.content.decode("utf-8")
            filename = slugify(link) + ".html"
            with open(f"wikipedia/{filename}", "w") as f:
                f.write(page_content)
    except Exception:
        pass

# Este código define una función llamada `index_page` que acepta un enlace como entrada.
# La función primero hace una solicitud GET al enlace. 
#Si la respuesta tiene un código de estado 200, significa que la solicitud se completó correctamente. 
#En ese caso, la función devuelve el contenido de la página en formato de cadena.

# La función `index_page` luego usa la función `slugify` de la biblioteca `slugify` 
#para crear un nombre de archivo para el archivo HTML que se va a crear. La función `slugify` toma una cadena como entrada 
#y devuelve una cadena que ha sido convertida a minúsculas y tiene todos los caracteres no ASCII reemplazados por guiones.

# Finalmente, la función `index_page` crea un archivo en la carpeta `wikipedia` con el nombre de archivo slugified 
# y escribe el contenido de la página en el archivo. Si se produce una excepción durante el proceso anterior, 
#la función simplemente pasa.

```

## Step 7: Perform the page indexing in parallel and note the difference in performance.

Remember to include `%%time` at the beginning of the cell so that it measures the time it takes for the cell to run.

```python
%%time
import multiprocessing

def index_page(link):
    try:
        response = requests.get(link)
        if response.status_code == 200:
            page_content = response.content.decode("utf-8")
            filename = slugify(link) + ".html"
            with open(f"wikipedia/{filename}", "w") as f:
                f.write(page_content)
    except Exception:
        pass

def download_wikipedia_pages_parallel(links):
    with multiprocessing.Pool(processes=3) as pool:
        pool.map(index_page, links)

if __name__ == "__main__":
    links = ["https://en.wikipedia.org/wiki/Python",
             "https://en.wikipedia.org/wiki/Machine_learning",
             "https://en.wikipedia.org/wiki/Deep_learning"]


    download_wikipedia_pages_parallel(links)
```
