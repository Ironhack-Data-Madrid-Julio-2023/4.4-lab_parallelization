## Step 2: Use BeautifulSoup to extract a list of all the unique links on the page.

```python
from bs4 import BeautifulSoup
import re
```

```python
#your code

url = 'https://en.wikipedia.org/wiki/Data_science'

wiki_data = requests.get(url)

soup = BeautifulSoup(wiki_data.content, 'html.parser')

links = soup.find_all('a', attrs={'title': True})

unique_links = set()
for link in links:
    href = link.get('href')
    if href:
        unique_links.add(href)

for link in unique_links:
    print(link)
```

```python
from urllib.parse import urljoin

links = soup.find_all('a', attrs={'title': True})

unique_links = set()
for link in links:
    href = link.get('href')
    if href:
        full_link = urljoin(url, href)
        
        if full_link.startswith('https://en.wikipedia.org/wiki/'):
            link_name = full_link.replace('https://en.wikipedia.org/wiki/', '')
            unique_links.add(f"https://en.wikipedia.org/wiki/{link_name}")

for link in unique_links:
    print(link)
```

## Step 3: Use list comprehensions with conditions to clean the link list.

There are two types of links, absolute and relative. Absolute links have the full URL and begin with http while relative links begin with a forward slash (/) and point to an internal page within the wikipedia.org domain. Clean the respective types of URLs as follows.

- Absolute Links: Create a list of these and remove any that contain a percentage sign (%).
- Relativel Links: Create a list of these, add the domain to the link so that you have the full URL, and remove any that contain a percentage sign (%).
- Combine the list of absolute and relative links and ensure there are no duplicates.

```python
url = 'http://wikipedia.org'
```

```python
#your code


# Send a GET request to the URL and retrieve the page content
response = requests.get(url)

# Create a BeautifulSoup object and parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find all anchor tags
links = soup.find_all('a')

# Lists to store absolute and relative links
absolute_links = []
relative_links = []

# Extract the links
for link in links:
    href = link.get('href')
    if href:
        # Make the links absolute by joining them with the base URL
        absolute_link = urljoin(url, href)
        absolute_links.append(absolute_link)

        # Check if the link is relative
        if not href.startswith('http'):
            relative_links.append(href)

# Print the absolute links
print("Absolute Links:")
for link in absolute_links:
    print(link)

print()

# Print the relative links
print("Relative Links:")
for link in relative_links:
    print(link)
```

```python
links = soup.find_all('a')

# Lists to store absolute and relative links
absolute_links = []
relative_links = []

# Extract the links
for link in links:
    href = link.get('href')
    if href:
        # Make the links absolute by joining them with the base URL
        absolute_link = urljoin(url, href)
        absolute_links.append(absolute_link)

        # Check if the link is relative
        if not href.startswith('http'):
            # Extract only the desired part of the link
            link_name = href.replace('/wiki/', '')
            # Construct the complete link
            full_link = f"https://en.wikipedia.org/wiki/{link_name}"
            relative_links.append(full_link)

# Print the absolute links
print("Absolute Links:")
for link in absolute_links:
    print(link)

print()

# Print the relative links
print("Relative Links:")
for link in relative_links:
    print(link)
```

```python
links = soup.find_all('a')

absolute_links = [urljoin(url, link.get('href')) for link in links if link.get('href')]

relative_links = [f"https://en.wikipedia.org/wiki/{link.get('href').replace('/wiki/', '')}" for link in links if link.get('href') and not link.get('href').startswith('http')]

total_list = list(set(absolute_links + relative_links))

for link in total_list:
    print(link)
```

## Step 5: Write a function called index_page that accepts a link and does the following.

- Tries to request the content of the page referenced by that link.
- Slugifies the filename using the `slugify` function from the [python-slugify](https://pypi.org/project/python-slugify/) library and adds a .html file extension.

    - You can then slugify a link as follows `slugify(link)`.
- Creates a file in the wikipedia folder using the slugified filename and writes the contents of the page to the file.
- If an exception occurs during the process above, just `pass`.

```python
%%time

#your code
url = 'https://en.wikipedia.org/wiki/Data_science'
folder_name = "wikipedia"


def index_page(url):
    
    try:
        response = requests.get(url)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"Error occurred while fetching {url}: {e}")
        return
    
    content = response.content

    soup = BeautifulSoup(content, 'html.parser')

    title = soup.title.string.strip()

    filename = "index.html"

    file_path = folder_name + "/" + filename

    with open(file_path, 'wb') as file:
        file.write(content)

    print(f"Page '{title}' is indexed and saved as '{filename}'.")

for relative_url in relative_links:
    url = 'https://en.wikipedia.org' + relative_url
    index_page(url)
```
