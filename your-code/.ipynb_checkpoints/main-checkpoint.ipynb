## Step 1: Use the requests library to retrieve the content from the URL below.

https://en.wikipedia.org/wiki/Data_science

```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/Data_science'

html = requests.get(url).content

html[0:600]
```

## Step 2: Use BeautifulSoup to extract a list of all the unique links on the page.

```python
import re

sopa = BeautifulSoup(html, "html.parser")


link = sopa.find_all('div', class_='mw-page-container-inner')
```

```python
etiquetas_a = sopa.find_all('a')
```

```python
valores_href = []

for etiqueta in etiquetas_a:
    if 'href' in etiqueta.attrs:
        valores_href.append(etiqueta['href'])
    else:
        valores_href.append(None)

print(valores_href)
```

## Step 3: Use list comprehensions with conditions to clean the link list.

There are two types of links, absolute and relative. Absolute links have the full URL and begin with http while relative links begin with a forward slash (/) and point to an internal page within the wikipedia.org domain. Clean the respective types of URLs as follows.

- Absolute Links: Create a list of these and remove any that contain a percentage sign (%).
- Relativel Links: Create a list of these, add the domain to the link so that you have the full URL, and remove any that contain a percentage sign (%).
- Combine the list of absolute and relative links and ensure there are no duplicates.

```python
domain = 'http://wikipedia.org'
html = requests.get(domain).content

html[0:600]
sopa = BeautifulSoup(html, "html.parser")
```

```python
#your code

link = sopa.find_all('div', class_='central-featured-lang')

link[0]
```

```python
etiquetas_a = sopa.find_all('a')

# Extraer el valor del atributo "href" de cada etiqueta "a" encontrada
valores_href = [etiqueta['href'] for etiqueta in etiquetas_a]

valores_href
```

```python
# Enlaces absolutos

#elimine los que contengan un signo de porcentaje (%) y empiece por http

enlaces_absolutos = [enlace for enlace in valores_href if enlace.startswith('http') and '%' not in enlace]

# Se utiliza.starwith para que sepa que son http

enlaces_absolutos
```

```python
# Enlaces relativos

# Se agrega el dominio al enlace para que tenga la URL completa y se elimina los que contengan un signo de porcentaje (%).
dominio = 'https://es.wikipedia.org' 
enlaces_relativos = [dominio + enlace for enlace in valores_href if enlace.startswith('/') and '%' not in enlace]
enlaces_relativos
```

```python
# Se agrupan los dos enlaces. 

enlaces_limpios = list(set(enlaces_absolutos + enlaces_relativos))
enlaces_limpios
```

## Step 5: Write a function called index_page that accepts a link and does the following.

- Tries to request the content of the page referenced by that link.
- Slugifies the filename using the `slugify` function from the [python-slugify](https://pypi.org/project/python-slugify/) library and adds a .html file extension.
    - If you don't already have the python-slugify library installed, you can pip install it as follows: `$ pip install python-slugify`.
    - To import the slugify function, you would do the following: `from slugify import slugify`.
    - You can then slugify a link as follows `slugify(link)`.
- Creates a file in the wikipedia folder using the slugified filename and writes the contents of the page to the file.
- If an exception occurs during the process above, just `pass`.

```python
#!pip install python-slugify
```

```python
from slugify import slugify

unicode=str()
```

```python
def index_page(s):
    
    try: #Intenta solicitar el contenido de la página a la que hace referencia ese enlace.

        #1- Slugifica el nombre del archivo y agrega una extensión de archivo .html.
        nombre_archivo = slugify(s) + '.html'
        
       
        #2- Solicitar el contenido de la pagina
        respuesta = requests.get(s)
        contenido = respuesta.text
        
        #3- Se crea un archivo en la carpeta wikipedia y escirbe el contenido.
        
        with open(nombre_archivo,"w") as archivo:
            archivo.write(contenido)
            

  # Se abre el archivo en modo escritura ("w") con codificación UTF-8 y escribimos el contenido de la página en el archivo.

    except:
        pass
    
```

```python
index_page('https://es.wikipedia.org//ja.wikipedia.org/')
```

## Step 6: Sequentially loop through the list of links, running the index_page function each time.

Remember to include `%%time` at the beginning of the cell so that it measures the time it takes for the cell to run.

```python
import time
```

```python
%%time

for enlace in enlaces_limpios:
    
    index_page(enlace)
    
```

## Step 7: Perform the page indexing in parallel and note the difference in performance.

Remember to include `%%time` at the beginning of the cell so that it measures the time it takes for the cell to run.

```python
import multiprocessing as mp
```

```python
%%time

pool = multiprocessing.Pool()

pool.map(index_page, enlaces_limpios)

pool.close()
pool.join()
```
